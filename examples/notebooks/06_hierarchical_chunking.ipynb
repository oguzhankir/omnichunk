# Hierarchical chunking

Multi-level `ChunkTree`: leaves for embeddings, roots for LLM context.

```python
import sys
from pathlib import Path

# Ensure the notebook imports local repo code (not an older pip wheel).
candidates = [
    Path.cwd() / "src",
    Path.cwd().parent / "src",
    Path.cwd().parent.parent / "src",
]
repo_src = next((p.resolve() for p in candidates if p.exists()), None)
if repo_src is None:
    raise RuntimeError("Could not locate local src/ directory for omnichunk")
if str(repo_src) not in sys.path:
    sys.path.insert(0, str(repo_src))

import omnichunk
from omnichunk import Chunker

print("omnichunk", omnichunk.__version__, "from", omnichunk.__file__)

src = 'def f():\n    return 1\n' * 5
c = Chunker(size_unit='chars', min_chunk_size=1)
tree = c.hierarchical_chunk('m.py', src, levels=[48, 200, 800])
print('levels', tree.level_count, 'nodes', len(tree.nodes))
print('leaves', len(tree.leaves()), 'roots', len(tree.roots())
```

```python
def ascii_tree(tree):
    roots = tree.roots()
    for ri, r in enumerate(roots):
        print(f'ROOT[{ri}] bytes {r.byte_range.start}-{r.byte_range.end}')
        for ch in tree.children(r):
            print(f'  └─ leaf bytes {ch.byte_range.start}-{ch.byte_range.end}')

ascii_tree(tree)
```
